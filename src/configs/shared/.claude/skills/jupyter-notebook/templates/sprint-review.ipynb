## Sprint Data

Paste or generate ticket data below. The `sprint-notebook.sh` routine
populates this from Jira automatically.

```python
# Sprint ticket data — populated by routine or manually
# Replace this with actual data from Jira
completed = pd.DataFrame({
    'key': [],
    'summary': [],
    'type': [],
    'priority': [],
    'story_points': [],
    'assignee': [],
    'resolved_date': [],
})

in_progress = pd.DataFrame({
    'key': [],
    'summary': [],
    'type': [],
    'priority': [],
    'status': [],
    'assignee': [],
})

print(f'Completed: {len(completed)} tickets')
print(f'In progress: {len(in_progress)} tickets')
```

## Completed Work

```python
if not completed.empty:
    display_cols = ['key', 'summary', 'type', 'priority', 'story_points']
    available = [c for c in display_cols if c in completed.columns]
    completed[available].style.set_caption('Completed Tickets')
else:
    print('No completed ticket data — populate the data cell above')
```

```python
# Status breakdown
total = len(completed) + len(in_progress)
if total > 0:
    status_counts = pd.DataFrame({
        'Status': ['Completed', 'In Progress'],
        'Count': [len(completed), len(in_progress)]
    })
    if HAS_PLOTLY:
        fig = px.pie(status_counts, values='Count', names='Status',
                     title='Sprint Status Breakdown',
                     color_discrete_sequence=['#27ae60', '#f39c12'])
        fig.show()
    else:
        fig, ax = plt.subplots(figsize=(8, 8))
        ax.pie(status_counts['Count'], labels=status_counts['Status'],
               autopct='%1.1f%%', colors=['#27ae60', '#f39c12'])
        ax.set_title('Sprint Status Breakdown')
        plt.show()
```

```python
# Tickets by type
if not completed.empty and 'type' in completed.columns:
    type_counts = completed['type'].value_counts()
    fig, ax = plt.subplots(figsize=(10, 6))
    type_counts.plot(kind='bar', ax=ax, color='steelblue')
    ax.set_title('Completed Tickets by Type', fontweight='bold')
    ax.set_ylabel('Count')
    ax.set_xlabel('')
    for container in ax.containers:
        ax.bar_label(container)
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()
```

## Velocity Trend

Populate `velocity_data` with historical sprint data to see the trend.

```python
# Historical velocity — populate with data from recent sprints
velocity_data = pd.DataFrame({
    'sprint': [],
    'completed': [],
    'points': [],
})

if not velocity_data.empty:
    fig, ax1 = plt.subplots(figsize=(12, 6))
    ax1.bar(velocity_data['sprint'], velocity_data['completed'],
            alpha=0.7, color='steelblue', label='Tickets')
    ax1.set_ylabel('Tickets Completed', color='steelblue')
    ax1.set_xlabel('Sprint')
    ax1.set_title('Sprint Velocity Trend', fontweight='bold')

    if velocity_data['points'].sum() > 0:
        ax2 = ax1.twinx()
        ax2.plot(velocity_data['sprint'], velocity_data['points'],
                 color='#e74c3c', marker='o', linewidth=2, label='Story Points')
        ax2.set_ylabel('Story Points', color='#e74c3c')

    plt.tight_layout()
    plt.show()
else:
    print('No velocity data — add historical sprint metrics above')
```
